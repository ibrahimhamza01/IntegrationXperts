```python
from ultralytics import YOLO
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
test_image = r"C:\Users\User\Downloads\UTKFace\12_0_0_20170103200900511.jpg.chip.jpg"
```

```python
model = YOLO(r"D:\IntegrationXpert\male, female & kids\yolo v8\runs\classify\train\weights\last.pt")  # load a custom model

test_image = Image.open(test_image)

# Predict on the test image
results = model(test_image)

# Get class names and probabilities
names_dict = results[0].names
probs = results[0].probs.data.tolist()

# Find the predicted class
predicted_class = names_dict[np.argmax(probs)]

# Plot the test image with the predicted class name
plt.imshow(test_image)
plt.title(f"Predicted: {predicted_class}")
plt.axis('off')
plt.show()
```
