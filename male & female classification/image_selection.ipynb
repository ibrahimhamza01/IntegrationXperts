```python
import os
import shutil
import random

def frame_skipping(source_dir, dest_dir, min_interval=4, max_interval=6):
    """Copy frames at random intervals to the destination directory."""
    if not os.path.exists(dest_dir):
        os.makedirs(dest_dir)

    files = sorted([f for f in os.listdir(source_dir) if f.lower().endswith(('.png', '.jpg', '.jpeg'))])
    i = 0
    while i < len(files):
        src_path = os.path.join(source_dir, files[i])
        dest_path = os.path.join(dest_dir, files[i])
        shutil.copyfile(src_path, dest_path)
        print(f"Copied frame: {dest_path}")

        # Generate a random skip interval for the next frame
        skip_interval = random.randint(min_interval, max_interval)
        i += skip_interval

# Paths to your source and destination directories
source_directory = r"D:\IntegrationXpert\male & female classification (1)\People dataset\video28"
destination_directory = r"D:\IntegrationXpert\male & female classification (2)\dataset\video28"

# Perform frame skipping with random intervals
frame_skipping(source_directory, destination_directory)
```

```python
import os
import shutil

def copy_and_rename_files(src_dir, dst_dir):
    # Ensure the destination directory exists
    os.makedirs(dst_dir, exist_ok=True)

    # Get a list of all files in the source directory
    files = os.listdir(src_dir)

    # Iterate over the files and copy them to the destination directory with new names
    for i, file_name in enumerate(files, start=231):
        src_file_path = os.path.join(src_dir, file_name)
        file_extension = os.path.splitext(file_name)[1]
        dst_file_name = f"image_{i}{file_extension}"
        dst_file_path = os.path.join(dst_dir, dst_file_name)
        
        # Copy the file
        shutil.copy(src_file_path, dst_file_path)
        print(f"Copied {file_name} to {dst_file_name}")

# Example usage
src_directory = r"D:\IntegrationXpert\male & female classification (2)\dataset\video32\labels"
dst_directory = r"D:\IntegrationXpert\male & female classification (2)\final dataset\labels"

copy_and_rename_files(src_directory, dst_directory)
```
